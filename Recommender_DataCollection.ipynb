## 1. Review Data

### 1.1. Start with Page 1

```python
url1 = 'http://www.basenotes.net/fragrancereviews/page/1'

headers1 = {
    'User-Agent': "Mozilla/5.0 (X11; Linux x86_64)\
    AppleWebKit/537.36 (KHTML, like Gecko)\
    Chrome/77.0.3865.90 Safari/537.36"
}

response1 = requests.get(url1, headers=headers1)

if response.status_code == 200:
    soup1 = BeautifulSoup(response1.content, 'html.parser')
else:
    print(response1.status_code)
```

```python
reviews1 = soup1.find_all('div', {'class': 'reviewmain'})
```

```python
reviews_list = []

for review in reviews1:
    # create a dictionary to store details for each review
        review_dict = {}

        # parse rating value
        rating_value = review['class'][-1][-1]

        # parse review id
        review_id = review['id'].split('-')[-1]

        # get user details
        user = review.find('div', {'class': 'reviewauthor'})

        # parse user id
        user_id = user.find('a')['href'].split('/')[-1]

        # parse user name
        user_name = user.text.split(' Show all reviews ')[0]

        # parse user location
        user_location = user.text.split(' Show all reviews ')[-1]

        # get review blurb details
        blurb = review.find('div', {'class': 'reviewblurb'})

        # parse scent id
        scent_id = blurb.find('a')['href'].split('ID')[-1].split('.')[0]

        # parse scent name
        scent_name = blurb.find('a').text.split(' by ')[0]

        # parse scent brand
        scent_brand = blurb.find('a').text.split(' by ')[-1]

        # parse review text
        review_text = blurb.text.split(scent_brand, 1)[-1]

        # add details to dictionary
        review_dict['rating_value'] = rating_value
        review_dict['review_id'] = review_id
        review_dict['user_id'] = user_id
        review_dict['user_name'] = user_name
        review_dict['scent_id'] = scent_id
        review_dict['scent_name'] = scent_name
        review_dict['scent_brand'] = scent_brand
        review_dict['review_text'] = review_text

        # if user has no location, return nan
        if user_location == '':
            review_dict['user_location'] = np.nan
        else:
            review_dict['user_location'] = user_location

        # append dictionary to the list
        reviews_list.append(review_dict)

# create a dataframe from the list of dictionaries
reviews_df = pd.DataFrame.from_dict(reviews_list)
```

# Goibibo Practice

```python
url = 'https://www.goibibo.com/hotels/hotels-in-shimla-ct/'

headers = {
    'User-Agent': "Mozilla/5.0 (X11; Linux x86_64)\
    AppleWebKit/537.36 (KHTML, like Gecko)\
    Chrome/77.0.3865.90 Safari/537.36"
}

response = requests.request('GET', url)

data = BeautifulSoup(response.text, 'html.parser')
# print(data)

cards_data = data.find_all('div', attrs={'class', 'HotelCardstyles__WrapperSectionMetaDiv-sc-1s80tyk-2 fKNNeH'})

print('Total number of cards found: ', len(cards_data))
```

```python
# create a list to store the data
scraped_data = []

for card in cards_data:
   
    # initialize the dictionary
    card_details = {}
    
    # get the hotel name
    hotel_name = card.find('div', attrs={'class': 'HotelCardstyles__HotelNameWrapperDiv-sc-1s80tyk-11 jkwhbV'})
    
    # get the room price
    room_price = card.find('span', attrs={'class': 'HotelCardstyles__CurrentPriceTextWrapper-sc-1s80tyk-24 dMzNlX'})
    
    # add data to the dictionary
    card_details['hotel_name'] = hotel_name.text
    card_details['room_price'] = room_price.text
    
    # append the scraped data to the list
    scraped_data.append(card_details)
    
# create a dataframe from the list of dictionaries
dataframe = pd.DataFrame.from_dict(scraped_data)

# save the scraped data as CSV file
dataframe.to_csv('hotels_data.csv', index=False)
```
